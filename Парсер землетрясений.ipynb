```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import folium
import re
```

```python
'''Паосим данные по координатам'''
URL = 'http://www.gsras.ru/cgi-bin/new/mapCustom.pl?l=0&lat=51.719&lon=94.43&num=1000&rad=400'
r = requests.get(URL)
df = pd.read_html(r.text, header=0)[2]
df['Время [GMT]'] = pd.to_datetime(df['Время [GMT]'])
p = re.compile(r'/')
df['mb'] = [p.sub('', x) for x in df['mb']]
df
```

```python
'''строим карту на основе folium'''
m = folium.Map(
    location=[51.719082, 94.433983],
    tiles='Stamen Toner',
    zoom_start=7
)
for er in df.values.tolist():
    st = (er[5])   
    lat = float(er[2])
    lot = float(er[3])    
    point = (lat,lot)
    folium.CircleMarker(
    location=point,
    radius=5,    
    color= '#cc',
    fill=True,
    fill_color='#ff0000'
).add_to(m)
m
```
